Quadratic Equation

```python
a = int(input("Enter value of a: "))
b = int(input("Enter value of b: "))
c = int(input("Enter value of c: "))
d = int((b*b)-(4*a*c))
e = 2*a
x1 = (-b+math.sqrt(d))/e
x2 = (-b-math.sqrt(d))/e
print("The result of the quadratic equation is x1 = ",x1," and x2 = ",x2)
```

Validity of Email Address

```python
import re
s = input("Enter email address: ")
if not (re.search("[a-zA-Z]*[0-9]*[_]*[.]*[0-9a-zA-Z]+@[a-zA-Z]+.[a-zA-Z]",s)):
  print("Invalid email address")
else:
  print("Valid email address")
```

Password verification

```python
import re
password = input("Enter password: ")
l = len(password)
if (l>=8):
  if not (re.search("[A-Z]",password)):
    print ("Uppercase character missing")
  elif not (re.search("[a-z]",password)):
    print ("Lowercase character missing")
  elif not (re.search("[0-9]",password)):
    print ("Numeric character missing")
  elif not (re.search("[#$%^&+=]",password)):
    print ("Special character missing")
  else:
    print ("Password Validated")
else:
  print("Length of password is less than 8 characters")
```
